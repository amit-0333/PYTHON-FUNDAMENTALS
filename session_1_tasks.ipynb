### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python

# Write your code here
import math

# Input for point 1
x1, y1 = map(int, input("Enter coordinates of point 1 (x1 y1): ").split())

# Input for point 2
x2, y2 = map(int, input("Enter coordinates of point 2 (x2 y2): ").split())

# Distance formula
d = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print("Distance:", d)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
# Input
head = int(input("Enter number of heads: "))
leg = int(input("Enter number of legs: "))

# Logic:
# dog + chicken = head
# 4*dog + 2*chicken = leg

# Solving equations:
# dog = (leg - 2*head) / 2
dog = (leg - 2 * head) // 2
chick = head - dog

# Output
print("Number of dogs:", dog)
print("Number of chickens:", chick)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
# Input first two terms and n
a1 = float(input("Enter first term (a1): "))
a2 = float(input("Enter second term (a2): "))
n = int(input("Enter N (term to find): "))

# Calculate common difference
d = a2 - a1

# Calculate Nth term
nth_term = a1 + (n - 1) * d

# Output result
print("Nth term of the arithmetic series is:", nth_term)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
# Input first fraction
num1 = int(input("Enter numerator of first fraction: "))
den1 = int(input("Enter denominator of first fraction: "))

# Input second fraction
num2 = int(input("Enter numerator of second fraction: "))
den2 = int(input("Enter denominator of second fraction: "))

# Sum of fractions
sum_num = (num1 * den2) + (num2 * den1)
sum_den = den1 * den2

# Output result
print("Sum of fractions:", sum_num, "/", sum_den)
````
