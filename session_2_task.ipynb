### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Write code here
salary = float(input("Enter your salary (CTC): "))

if salary <= 500000:
    print("No tax for you:", salary)

elif salary <= 1000000:
    hra = salary * 0.10
    da = salary * 0.05
    pf = salary * 0.03

    tax = salary * 0.10

    final_salary = salary - (hra + da + pf + tax)
    print("In-hand salary:", final_salary)

elif salary <= 2000000:
    hra = salary * 0.10
    da = salary * 0.05
    pf = salary * 0.03

    tax = salary * 0.20

    final_salary = salary - (hra + da + pf + tax)
    print("In-hand salary:", final_salary)

else:
    hra = salary * 0.10
    da = salary * 0.05
    pf = salary * 0.03

    tax = salary * 0.30

    final_salary = salary - (hra + da + pf + tax)
    print("In-hand salary:", final_salary)
````

### `Problem 5` - Exercise 12: Display Fibonacci series up to 10 terms.

*Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34*

````python
# Write code here
# Number of terms
n = 10

# First two terms
a = 0
b = 1

# Print Fibonacci series
print("Fibonacci series up to 10 terms:")

for i in range(n):
    print(a, end=" ")

    # Next term
    c = a + b
    a = b
    b = c
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
# Write code here
# Input number
n = int(input("Enter a number: "))

# Initialize factorial
fact = 1

# Loop to calculate factorial
for i in range(1, n + 1):
    fact = fact * i

# Output result
print("Factorial of", n, "is:", fact)
````

### `Problem 7` - Reverse a given integer number.

Example:

`Input:`

```bash
76542
```

`Output:`

```bash
24567
```

````python
# Write code here
# Input number
num = int(input("Enter a number: "))

rev = 0
temp = num  # store original number

# Reverse logic
while temp > 0:
    digit = temp % 10        # get last digit
    rev = rev * 10 + digit   # build reversed number
    temp = temp // 10        # remove last digit

print("Reversed number:", rev)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Write code here
# Input
n = int(input("Enter value of N: "))

i = 1
total = 0

# Using while loop (no for loop allowed)
while i <= n:

    # Skip numbers divisible by 5
    if i % 5 == 0:
        i += 1
        continue

    total += i

    # Stop if sum exceeds 300
    if total > 300:
        break

    i += 1

print("Final sum:", total)
````

### `Problem 9`: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

````python
# Write code here
count = 0
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))

    if num == 0:
        break

    total += num
    count += 1

# Avoid division by zero
if count == 0:
    print("No numbers were entered.")
else:
    average = total / count
    print("Sum =", total)
    print("Average =", average)
````

###`Problem 9`: Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

````python
# Write code here
result = []

for i in range(2000, 3201):
    if i % 7 == 0 and i % 5 != 0:
        result.append(str(i))

print(",".join(result))
````

###`Problem 10`: Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.



````python
# Write code here

result = []

for num in range(1000, 3001):
    s = str(num)

    # check if all digits are even
    if all(int(digit) % 2 == 0 for digit in s):
        result.append(s)

print(" ".join(result))
````

###`Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# Write code here
import math

x = 0
y = 0

while True:
    move = input().strip()

    if move == "!":
        break

    direction, steps = move.split()
    steps = int(steps)

    if direction == "UP":
        y += steps
    elif direction == "DOWN":
        y -= steps
    elif direction == "LEFT":
        x -= steps
    elif direction == "RIGHT":
        x += steps

# Calculate distance from origin
distance = math.sqrt(x**2 + y**2)

# Print nearest integer
print(round(distance))
````

###`Problem 12`:Write a program to print whether a given number is a prime number or not

````python
# Write code here
# Input number
n = int(input("Enter a number: "))

# Prime check
if n <= 1:
    print("Not a Prime Number")
else:
    is_prime = True

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime Number")
    else:
        print("Not a Prime Number")
````

###`Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# Write code here
# Input range from user
start = int(input("Enter starting range: "))
end = int(input("Enter ending range: "))

print("Armstrong numbers are:")

for num in range(start, end + 1):
    temp = num
    total = 0

    while temp > 0:
        digit = temp % 10
        total += digit ** 3
        temp = temp // 10

    if total == num:
        print(num, end=" ")
````

###`Problem 14`:Calculate the angle between the hour hand and minute hand.

Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final result angle. For example, if the final angle is 10.61, we need to print 10.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

````python
# Write code here
import math

# Input
H = int(input("Enter hour: "))
M = int(input("Enter minutes: "))

# Convert to 12-hour format
H = H % 12

# Calculate angles
hour_angle = (30 * H) + (0.5 * M)
minute_angle = 6 * M

# Difference
angle = abs(hour_angle - minute_angle)

# Minimum angle
min_angle = min(angle, 360 - angle)

# Floor value
print(math.floor(min_angle))
````

###`Problem 15`:Given two rectangles, find if the given two rectangles overlap or not. A rectangle is denoted by providing the x and y coordinates of two points: the left top corner and the right bottom corner of the rectangle. Two rectangles sharing a side are considered overlapping. (L1 and R1 are the extreme points of the first rectangle and L2 and R2 are the extreme points of the second rectangle).

Note: It may be assumed that the rectangles are parallel to the coordinate axis.

<img src='https://www.geeksforgeeks.org/wp-content/uploads/rectanglesOverlap.png' width='300' height='200'>

````python
# Write code here
# Input for rectangle 1
x1 = int(input("Enter L1 x: "))
y1 = int(input("Enter L1 y: "))
x2 = int(input("Enter R1 x: "))
y2 = int(input("Enter R1 y: "))

# Input for rectangle 2
x3 = int(input("Enter L2 x: "))
y3 = int(input("Enter L2 y: "))
x4 = int(input("Enter R2 x: "))
y4 = int(input("Enter R2 y: "))

# Check overlap condition
if x2 < x3 or x4 < x1 or y2 < y3 or y4 < y1:
    print("Rectangles do NOT overlap")
else:
    print("Rectangles overlap")
````
